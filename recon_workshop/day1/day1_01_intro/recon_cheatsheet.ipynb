# Recon - Cheatsheet

## What do we try to look for when performing reconnaissance aka information gathering?

Everything we search for, our aim is to increase our attack surface, so we can narrow down to interesting results.

### Organisation

Profiling an Organisation

- company careers
    - what kind of technologies or people are they looking for?
- acquisitions and mergers
    - are the new investments secure?
- people connected to the company
    - high profile
    - different departments
    - what are they posting about the company?
- news
    - can lead to more actionable intelligence
    - who is mentioned? any tech?



### Technology

- subdomains, endpoints
    - other people might be doing bug bounty and have their lists online
- internal development
    - .int, .internal, .local, more...
- sensitive information leaks such as credentials, sub-domains and more
    - be creative
    - depending on the technology, there could be naming conventions - Google it! Look at official documentation
    - credentials, config, password, token, secret, key, login, ftp, pass, admin, pwd, more...
    - ssh2_auth_passwd -> Unauthorized access to servers
- public repos that should be private
- services 
    - ssh, ftp, sftp, vpn, http, vsphere, siem, firewall, citrix, fortinet
- ports
    - 20, 21, 23, 25, 53, 80, 110, 111, 135, 139, 143, 443, 445, 993, 995, 1723, 3306, 3389, 5900, 8008, 8080, 8443
    - so many can be listed here, and services won't always be configured on default ports
    - nice reference: https://www.speedguide.net/ports.php
- third parties the company uses
- employees
- company infrastructure
