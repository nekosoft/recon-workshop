### Certificate Transparency

The little green lock is a certificate that can serve as a proof of identity, so your browser knows it's talking to the right server. Certificates are issued by Certificate Authorities - transparency allows people to look up the information of these certificates to make sure a website is legitimate. 
Transparency reason : https://threatpost.com/final-report-diginotar-hack-shows-total-compromise-ca-servers-103112/77170/

- Certificates: https://crt.sh/
- sslmate
