```python
from datetime import datetime

# Create a datetime string to be used for saving our file
dt = datetime.now()
dt_string = dt.strftime("%Y-%m-%d-%H%M%S")

# Return functions

# Return the name as is, so we can further process it
# Input: A name, example: 'john'
# Outpit: The same name, example: 'john'
def return_name_unmodified(a):
    return a

# Return only the first character of a name
# Input: A name, example: a = 'john'
# Output: The first letter, example: 'j'
def return_first_character(a):
    return a[0]

# An array of functions, so we can loop through each of them
return_unmodified_names_functions = [
    return_name_unmodified,
    return_first_character
]


# Merge functions

# Merge two names
# Input: Two names, example, a = 'john', b = 'veem'
# Output: 'johnveem'
def merge_names(a, b):
    return a + b

# Merge two names with a period
# Input: Two names, example a = 'john, b = 'veem'
# Output: 'john.veem'
def merge_names_period(a, b):
    return a + '.' + b

# An array of functions, so we can loop through them
merge_names_functions = [
    merge_names,
    merge_names_period
]

# A function that creates email patterns programmatically
# Parameters: 
# first_name
# second_name
# names_array = an array to store email patterns
def name_permutations(first_name, second_name, names_array):
    
    # Loop through the return_unmodified_names_functions to simply 
    # store the first part of the email in a variable
    for function in return_unmodified_names_functions:
        first_part_of_email = function(first_name)
        # print("First part: ", first_part_of_email)
        
        # Loop through the return_unmodified_names_functions to
        # store the second part of the email in a variable
        for function in return_unmodified_names_functions:
            second_part_of_email = function(second_name)
            
            # Loop through the merge_names_functions to
            # programatically build email patterns
            for merge_function in merge_names_functions:
                
                # Add email pattern names to build the array
                names_array.append(merge_function(first_part_of_email, second_part_of_email))
                names_array.append(merge_function(second_part_of_email, first_part_of_email))

permutations = []

# you can provide your own txt file of names
with open('names.txt', 'r') as file:
    for index, line in enumerate(file):
        names = line.lower().split()
        name_permutations(names[0], names[1], permutations)
        # name_permutations(names[1], names[0], permutations)
        # print(name_permutations(names[0], names[1]))
        # print(name_permutations(names[1], names[0]))
        
print(permutations)

# talk about append - a - instead of write - w

# The filename includes a date time string so that you don't accidentally
# overwrite the file

# What happens if the filename stayed the same?
filename = 'permutated_names_' + dt_string + '.txt'

with open(filename, 'w') as outfile:
    for name in permutations:
        outfile.write(name + '@example.com\n') # you can change this domain name
```
